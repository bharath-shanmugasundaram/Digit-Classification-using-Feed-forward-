```python
import numpy as np
import tensorflow as tf
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
def equalizer(n):
    lis = np.zeros(10, dtype=int)
    lis[n] = 1
    return lis
```

```python

X = np.array([i.flatten() for i in X_train])
Y = np.array([equalizer(i) for i in y_train])
X = X / 255.0

W = np.random.randn(784, 10) * 0.01
b = np.zeros((1, 10))
```

```python
def softmax(z):
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
    print(z)
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

def cross_entropy(y_true, y_pred):
    m = y_true.shape[0]
    return -np.sum(y_true * np.log(y_pred + 1e-9)) / m
```

```python
epochs = 10000
alpha = 0.01

for epoch in range(epochs):
    z = np.dot(X, W) + b
    y_pred = softmax(z)
    loss = cross_entropy(Y, y_pred)

    dz = y_pred - Y
    dW = (X.T @ dz) / X.shape[0]
    db = np.sum(dz, axis=0, keepdims=True) / X.shape[0]

    W -= alpha * dW
    b -= alpha * db

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")
```

```python
import numpy as np

# Flatten and normalize test data
X_test_flat = X_test.reshape(X_test.shape[0], -1) / 255.0   # shape: (10000, 784)

# If y_test is one-hot encoded, convert it
if len(y_test.shape) > 1 and y_test.shape[1] > 1:
    y_test = np.argmax(y_test, axis=1)

# --- Softmax ---
def softmax(z):
    z = z - np.max(z, axis=1, keepdims=True)  # for numerical stability
    exp_z = np.exp(z)
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

# --- Prediction ---
def predict(X, W, b):
    z = np.dot(X, W) + b
    probs = softmax(z)
    return np.argmax(probs, axis=1)

# --- Accuracy ---
def accuracy(X, y, W, b):
    y_pred = predict(X, W, b)
    return np.mean(y_pred == y)

# Test accuracy
acc = accuracy(X_test_flat, y_test, W, b)
print(f"Test Accuracy: {acc * 100:.2f}%")
```
